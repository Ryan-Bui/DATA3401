# Matplotlib


```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0,5,10)
y = x**2

plt.subplot(1,2,1) # number of rows, number cols, position of this plot
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
plt.show
```

```python
fig, axes = plt.subplots(nrows=1, ncols = 2)
print(axes)
print(type(fig))
print(type(axes))
```

```python
fig, axes = plt.subplots(nrows=1, ncols = 2)
for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Title')
```

```python
fig, axes = plt.subplots(1,3, figsize=(15,4))

axes[0].plot(x,x**2,x,x**3)
axes[0].set_title('Default axes ranges')

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title('Tight axes')

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title('Custom axes range')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x,x**2, label='y = x**2')
ax.plot(x,x**3, label='y = x**3')
ax.legend(loc=0)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x,x+1,color='red',alpha=0.5)
ax.plot(x,x+2,color='#1155dd')
ax.plot(x,x+2,color='#15cc55')
plt.show()
        
```

```python
fig, ax = plt.subplots(figsize=(10,4))

ax.plot(x,x**2,x,x**3,lw=2)
ax.set_xticks([1,2,3,4,5])

ax.set_yticks([0,50,100,150])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,4))

ax.plot(x,x**2,x,x**3,lw=2)
ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels(['a','b','c','d','e'], fontsize = 16)

ax.set_yticks([0,50,100,150])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,4))

ax.plot(x,x**2,x,x**3,lw=2)
ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels(['a','b','c','d','e'], fontsize = 16)

yticks=[1,5,100,150]
ax.set_yticks(yticks)
ax.set_yticks()
plt.show()

```

```python
fig = plt.figure()
ax = fig.add_axes([0.0,0.0,0.6,0.6],polar=True)
t = np.linspace(0,2*np.pi,100)
ax.plot(t,np.cos(t),color='blue',lw=3)
plt.show()
```
