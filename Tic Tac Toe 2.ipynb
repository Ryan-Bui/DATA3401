```python
def create_game_board(size=3):
    board = [[empty]*size for i in range(size)]
    return board
```

```python
board_character = { player_1: player_1_piece,
                    player_2: player_2_piece,
                    empty: empty_space }

row_names=list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
row_map=dict(zip(row_names,range(size)))

column_names=list(map(str,range(1,size+1)))
column_map=dict(zip(column_names,range(size)))

```

```python
 #checking rows(p1)
def check_rows(board):
    for i in range(size):
        win_1 = True
        for j in range(size):
            if board[i][j] != player_1:
                win_1 = False
                break
        if win_1:
            return True
    for i in range(size):
        win_2 = True
        for j in range(size):
            if board[i][j] != player_2:
                win_2 = False
                break
        if win_2:
            return True


    # checking columns(p1)
def check_cols(board):
    for i in range(size):
        win_1 = True
        for j in range(size):
            if board[j][i] != player_1:
                win_1 = False
                break
        if win_1:
            return True
    for i in range(size):
        win_2 = True
        for j in range(size):
            if board[j][i] != player_2:
                win_2 = False
                break
        if win_2:
            return True

        # checking diagonals(p1)
def check_diag(board):
    win_1 = True
    for i in range(size):
        if board[i][i] != player_1:
            win_1 = False
            break
    for i in range(size):
        if board[i][size - 1 - i] != player_1:
            win_1 = False
            break
    if win_1:
        return True

    
    win_2 = True
    for i in range(size):
        if board[i][i] != player_2:
            win_2 = False
            break
    if win_2:
        return True

    for i in range(size):
        if board[i][size - 1 - i] != player_2:
            win_2 = False
            break
    if win_2:
        return True
    

def check_empty(board):
    for i in range(size):
        if board[i] != empty:
            return -1
        
def check_full(board):
    temp = 0
    for i in range(size):
        for j in range(size):
            if board[i][j] != empty:
                temp+=1
    if temp==(size**2):
        return 0
    else:
        return -1

def check_win(board):
    row_state = check_rows(board)
    col_state = check_cols(board)
    diag_state = check_diag(board)
    empty_state = check_empty(board)
    full_state = check_full(board)
    if row_state == True:
        return True
    elif col_state == True:
        return True
    elif diag_state == True:
        return True
    elif full_state == 0:
        return True
    else:
        return False
       # return True
   # else:
       # return False

    
        
```

```python
def create_piece(board,player,location,verbose=True):
    x,y = location
    move_possible = x < size and \
                    x >= 0 and \
                    y < size and \
                    y >= 0
    if not (move_possible):
        print_message('Move is off the board.',verbose)
        return False
    if board[x][y] == player:
        print("A piece is already there try again.")
        return False

    
    if move_possible and board[x][y] == empty:
        board[x][y] = player
        print_message("Player{0}'s piece has been placed.".format(player),verbose)
        return True
    else:
        print_message('Move is invalid.',verbose)
        return False

    
def print_message(message, verbose=True):
    if verbose:
        print(message)


def parse_location(location_string):
    
    if not isinstance(location_string,str):
        print_message("Invalid input. Location must be a string.")
        return False
    
    if len(location_string)!=2:
        print_message("Inavlid input. Location must be 2 characters")
        return False
    
    row = location_string[0].upper()
    col = location_string[1].upper()
    
    if not row in row_map:
        print_message("Not a valid row.")
        return False
    
    if not col in column_map:
        print_message("Not a valid column.")
        return False
    
    return row_map[row], column_map[col]

def nice_move_piece(board,player,location_string):
    location = parse_location(location_string)
    if location:
        return create_piece(board,player,location)
    else:
        print("Invalid Move")
        return False
```

```python
import copy

Possible_moves = ['A1','A2','A3','B1','B2','B3','C1','C2','C3']
def generate_moves(board,player,current_player,depth, keep_all_moves=False):
    if depth==0:
        return list(),score_board(board,player)

    moves=list()
    scores=list()
    
    # Look through the board for pieces belonging to a specific player
    for i in range(size):
        for j in range(size):
            if board[i][j]==current_player:
                
                # For each piece, consider moving left or right
                for move in [Possible_moves]:
                    
                    # Create a new copy of the board
                    new_board=copy.deepcopy(board)
                    
                    # Check if its a valid move
                    if create_piece(new_board,current_player,(i,j),move,verbose=False):
                        # If so, generate subsequent moves
                        next_moves,score=generate_moves(new_board,player,
                                                        switch_player(current_player),
                                                        depth-1,
                                                        keep_all_moves=keep_all_moves)
                        
                        # Keep track of this move
                        this_move=[(current_player,(i,j),move)]
                        if keep_all_moves:
                            # For our purposes, we usually only need to keep the top move!
                            this_move.extend(next_moves)
                        moves.append(this_move)
                        scores.append(score)

    return moves,scores
```
